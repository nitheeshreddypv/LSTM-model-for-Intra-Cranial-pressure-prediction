```python
import pandas as pd
import numpy as np

# Load the raw ICP data into a pandas dataframe
df = pd.read_csv('charis1.csv')

# Set a threshold range for reasonable ICP data
lower_threshold = -5
upper_threshold = 50

# Remove any data points outside of the threshold range
df.loc[df['ICP'] < lower_threshold, 'ICP'] = np.nan
df.loc[df['ICP'] > upper_threshold, 'ICP'] = np.nan

# Forward fill any missing data points to maintain data continuity
df.fillna(method='ffill', inplace=True)

# Define the smoothing window size (in number of data points)
window_size = 60000

# Define the step size for moving the smoothing window
step_size = int(window_size * 0.05)

# Initialize the start and end indices of the smoothing window
start_index = 0
end_index = window_size

# Loop over the data and apply the smoothing algorithm
while end_index <= len(df):
    # Calculate the mean and standard deviation of the data within the window
    window_mean = df['ICP'][start_index:end_index].mean()
    window_std = df['ICP'][start_index:end_index].std()

    # Find any outliers (values outside of +/-3 standard deviations) within the window
    outliers = np.where((df['ICP'][start_index:end_index] < window_mean - 3 * window_std) | (df['ICP'][start_index:end_index] > window_mean + 3 * window_std))

    # Replace outliers with the window mean
    df.loc[start_index + outliers[0], 'ICP'] = window_mean

    # Move the window forward by the step size
    start_index += step_size
    end_index += step_size

# Compress the data to 60 seconds per sample (0.017 Hz)
compressed_data = df.groupby(np.arange(len(df))//3000).mean()


```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,mean_squared_error

# Assume the actual and predicted values are stored in actual_data and predicted_data variables respectively
# Convert actual_data and predicted_data to binary labels (0 or 1) using a threshold of 0.5

# Calculate accuracy
accuracy = accuracy_score(actual_events, predicted_events)
c=confusion_matrix(actual_events, predicted_events)
print(c)
# Calculate confusion matrix
tn, fp, fn, tp = confusion_matrix(actual_events, predicted_events).ravel()

# Calculate specificity
specificity = tn / (tn + fp)

# Calculate sensitivity
sensitivity = tp / (tp + fn)
# RMSE
rmse = np.sqrt(mean_squared_error(actual_events, predicted_events))

# Print the results
print('Accuracy:', accuracy)
print('Specificity:', specificity)
print('Sensitivity:', sensitivity)
print('RMSE:', rmse)

```
